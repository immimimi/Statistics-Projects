```python
import numpy as np
import matplotlib.pyplot as plt
#finding the local minimum of a function for 3D
#define z func

def z_func(x,y):
    return np.sin(5 * x) * np.cos(5 * y) /5

def calculate_gradient(x, y): #we have our two partial derivitives to find gradient
    return np.cos(5 * x) * np.cos(5 * y), -np.sin(5 * x) * np.sin(5 * y)

x = np.arange(-1,1,0.05)
y = np.arange(-1,1,0.05)

#turn into a meshgrid
X, Y = np.meshgrid(x,y)

Z = z_func(X,Y)

current_pos = (0.7,0.4,z_func(0.7,0.4))

learning_rate = 0.01

ax = plt.subplot(projection="3d", computed_zorder=False) #zorder prioritizes the check point, so we can always see it on the graph


for _ in range(1000): # usually you have a tolerace instead of a fixed number, so it stops within a close enough range of the minimum

    #calculate positions
    X_derivative , Y_derivative = calculate_gradient(current_pos[0], current_pos[1])
    X_new, Y_new = current_pos[0] - learning_rate * X_derivative, current_pos[1] - learning_rate * Y_derivative
    current_pos = (X_new, Y_new , z_func(X_new,Y_new))

    #ploting

    ax.plot_surface(X,Y,Z, cmap="viridis", zorder = 0)
    ax.scatter(current_pos[0], current_pos[1], current_pos[2], color = "magenta", zorder=1)
    plt.pause(0.001)
    ax.clear()
```
